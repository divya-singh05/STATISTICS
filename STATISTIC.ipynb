```python
Q1=sub_df['Revenues'].quantile(0.25)
Q1
```

```python
Q3=sub_df['Revenues'].quantile(0.75)
Q3
```

```python
#IQR=Q3-Q1

IQR0=Q3-Q1
IQR0
```

```python
#UPPER LIMIT

upper_limit= Q3+1.5*IQR0
upper_limit
```

```python
#LOWER LIMIT

lower_limit= Q1-1.5*IQR0
lower_limit
```

```python
#OUTLINERS

outliers=sub_df[(sub_df['Revenues']<lower_limit)|(sub_df['Revenues']>upper_limit)]
print('Outliers rows:')
print(outliers)
```
